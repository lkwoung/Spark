```
#Pandas Dataframe 생성
import numpy as np
import pandas as pd
from sklearn.datasets import load_wine

wine = load_wine()

pdf_wine = pd.DataFrame(data= np.c_[wine['data'], wine['target']], columns= wine['feature_names'] + ['target'])
pdf_wine
```

```
#Spark Dataframe 생성 (Pandas Dataframe 이용)
from pyspark.sql import SparkSession

spark = SparkSession.builder.getOrCreate()
sdf_wine = spark.createDataFrame(pdf_wine)
sdf_wine.show()
```

```
# Pandas Dataframe Magnesium feature 최대/최소 값

#print(pdf_wine.max())
#print(pdf_wine.min())

pdf_wine_drop=pdf_wine.dropna(axis=0) #결측값 제거
pdf_magnesium=pdf_wine_drop.magnesium #특정 열 추출

pdf_max=pdf_magnesium.max()
pdf_min=pdf_magnesium.min()

print("pdf_max_magnesium=", pdf_max, "pdf_min_magnesium=", pdf_min)
```

```
# Spark Dataframe Magnesium feature 최대/최소 값
sdf_wine_drop=sdf_wine.dropna('any') #결측값 제거

sdf_max=sdf_wine_drop.agg({"magnesium": "max"}).collect()
sdf_min=sdf_wine_drop.agg({"magnesium": "min"}).collect()
print(sdf_max, sdf_min)
```

```
# Target으로 그룹한 후에 Pandas에서 Magnesium feature의 최대/최소 값
pdf_max2=pdf_wine_drop.groupby('target').max().magnesium.max()
pdf_min2=pdf_wine_drop.groupby('target').min().magnesium.min()

print("pdf_max_magnesim=", pdf_max2, "pdf_min_magnesim=", pdf_min2)
```

```
# Target으로 그룹한 후에 Spark에서 Magnesium feature의 최대/최소 값
sdf_wine_drop.groupBy("target").max("magnesium").agg({"max(magnesium)": "max"}).show()
sdf_wine_drop.groupBy("target").min("magnesium").agg({"min(magnesium)": "min"}).show()
```
